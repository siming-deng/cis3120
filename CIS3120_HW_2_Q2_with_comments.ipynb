```python
import requests
import json
import pandas as pd
```

```python
# key obtained from weather API
key = '8c8e0a888b14ee4390a4d755988e34a4'

cities = ['Seattle', 'Boston', 'Philadelphia', 'Washington, D.C.', 'Miami', 'Las Vegas', 
          'Baltimore', 'Atlanta', 'New York', 'Chicago', 'San Francisco', 'Los Angeles']

# created lists to store the values
feels = []
humidity = []
temp = []
max_temp = []
min_temp = []
weather = []
wind = []

# loop over each city in the cities list
for city in cities:
  # sending request using the url
  url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&APPID={key}&units=imperial'
  r = requests.get(url)

  # if the request was successful, print out the statments and append the value to the list
  if (r.status_code == 200):
    data = r.json()
    print(f"{data['name']} feels like {data['main']['feels_like']} °F.")
    feels.append(data['main']['feels_like'])
    print(f"{data['name']} humidity is {data['main']['humidity']} %.")
    humidity.append(data['main']['humidity'])
    print(f"{data['name']} temperature is {data['main']['temp']} °F.")
    temp.append(data['main']['temp'])
    print(f"{data['name']} maximum temperature is {data['main']['temp_max']} °F.")
    max_temp.append(data['main']['temp_max'])
    print(f"{data['name']} minimum temperature is {data['main']['temp_min']} °F.")
    min_temp.append(data['main']['temp_min'])
    print(f"{data['name']} weather description is {data['weather'][0]['description']}.")
    weather.append(data['weather'][0]['description'])
    print(f"{data['name']} wind speed is {data['wind']['speed']} mph.")
    wind.append(data['wind']['speed'])
    print(" ")
  else:
    # if the request was not successful, returning the unsuccessful message.
    print("API request was unsuccessful.")

# creating the series
weathers_data = {
    'City': cities,
    'Feels Like': feels,
    'Humidity': humidity,
    'Temperature': temp,
    'Maximum Temperature': max_temp,
    'Minimum Temperature': min_temp,
    'Weather Description': weather,
    'Wind Speed': wind
}

# using the series to create data frame
weather_df = pd.DataFrame(weathers_data)
```
